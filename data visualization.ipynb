```python
#Coversion of iris data into 3 dimentional data with 3D chart
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
x=PCA(n_components=3).fit_transform(iris.data)
y=iris.target
fig=plt.figure(1,figsize=(8,6))
ax=Axes3D(fig,elev=-150,azim=110)
ax.scatter(x[:, 0],x[:, 1],x[:, 2],c=y,cmap=plt.cm.rainbow_r)
ax.set_title("3D Chart")
ax.set_xlabel("Dimension one X")
ax.set_ylabel("Dimension two Y")
ax.set_zlabel("Dimension Three Z")
```
